```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/filtered_agrofood.csv")
df
```

#Define category from Agrofood emission sources

```python
# Define category mappings
categories = {
    "Crop & Soil Management": ["Crop Residues", "Rice Cultivation", "Drained organic soils (CO2)"],
    "Livestock & Manure Management": ["Manure applied to Soils", "Manure left on Pasture", "Manure Management"],
    "Agricultural Inputs & Manufacturing": ["Fertilizers Manufacturing", "Pesticides Manufacturing", "IPPU"],
    "Food Processing & Distribution": ["Food Processing", "Food Packaging", "Food Transport"],
    "Consumer & Waste Management": ["Food Retail", "Food Household Consumption", "Agrifood Systems Waste Disposal"]
}
```

```python

# Create a dictionary to store categorized datasets
category_dfs = {}

# Iterate through each category and create a new dataset
for category, columns in categories.items():
    # Select relevant columns along with 'Year', 'Area', and 'region'
    category_df = df[['Year', 'Area', 'region'] + columns].copy()

    # Add a 'Category' column
    category_df['Category'] = category

    # Store the filtered dataframe
    category_dfs[category] = category_df

# Display the first few rows of each category dataset
for category, category_df in category_dfs.items():
    print(f"\nData for category: {category}")
    print(category_df.head())
```

#Agricultural Inputs & Manufacturing categoty (top emission)

```python
category_dfs["Agricultural Inputs & Manufacturing"]
```

```python
# Set Seaborn style
sns.set_style("whitegrid")


# Extract the dataset for the selected category
agri_inputs_df = category_dfs["Agricultural Inputs & Manufacturing"]

# Group by Year and sum emissions across all relevant columns
agri_inputs_trend = agri_inputs_df.groupby("Year").sum(numeric_only=True).reset_index()

# Convert emissions from kilotonnes (kt) to Billion Metric Tonnes (BMT)
conversion_factor = 1e-6
numeric_cols = categories["Agricultural Inputs & Manufacturing"]
agri_inputs_trend[numeric_cols] *= conversion_factor  # Apply conversion

# Plot the line graph
plt.figure(figsize=(10, 6))
for column in numeric_cols:
    if column in agri_inputs_trend.columns:  # Ensure column exists before plotting
        plt.plot(agri_inputs_trend["Year"], agri_inputs_trend[column], marker=".", label=column)

# Customize plot
plt.xlabel("Year", fontsize=12)
plt.ylabel("Emissions (Billion Metric Tonnes CO₂)", fontsize=12)
plt.title("Agricultural Inputs & Manufacturing CO2 Emissions Over Time", fontsize=14, fontweight="bold")
plt.legend(title="Emission Sources", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.xticks(rotation=0)

# Show the plot
plt.show()
```
